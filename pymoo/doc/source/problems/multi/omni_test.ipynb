# Omni-test
The Omni-test problem is a multi-modal multi-objective optimization problem proposed by Deb in []. It has two objective
functions. Suppose that the dimension of the decision space is $D$, then it has $3^D$ Pareto subsets in the decision
space corresponding to the same Pareto front.

```python
from pymoo.problems.multi.omnitest import OmniTest
from pymoo.visualization.scatter import Scatter
```

## 2-dimensional case
### Pareto front

```python
problem = OmniTest(n_var=2)
pareto_front = problem.pareto_front(1000)
Scatter(title="Pareto front").add(pareto_front).show()
```

### Pareto set

```python
pareto_set = problem.pareto_set(1000)
Scatter(title="Pareto set").add(pareto_set).show()
```

## 3-dimensional case
### Pareto front

```python
problem = OmniTest(n_var=3)
pareto_front = problem.pareto_front(3000)
Scatter(title="Pareto front").add(pareto_front).show()
```

### Pareto set

```python
# noinspection PyUnresolvedReferences
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)
ax.view_init(elev=20, azim=5)
ax.set(title="Pareto set", xlabel="$x_1$", ylabel="$x_2$", zlabel="$x_3$")
ax.scatter(pareto_set[:, 0], pareto_set[:, 1], pareto_set[:, 2])
plt.show()
```
